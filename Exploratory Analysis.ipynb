```python
import pandas as pd
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline 
sns.set(color_codes=True)
```

```python
df=pd.read_csv("/Users/user/Documents/Salary.csv")
df.shape
```

```python
df.columns
```

```python
#remove duplicates, 88 observations left
df=df.drop_duplicates(subset=('Job Title', 'Company Name','Salary'))
df
```

```python
#deleting useless columns
df=df.drop(["Headquarters","Unnamed: 0","Competitors"],axis=1)
df.head()
```

```python
df.Salary.hist()
df['Salary'].mean()
#mean salary
```
